## Reading the dataset

```python
# importing necessary module.
import pandas as pd
# reading the dataset.
books = pd.read_csv("books.csv")
books.head()
```

# "Perform data cleaning and preprocessing steps to ensure the dataset is ready for analysis."

## Checking for nulls.

```python
print(books.isnull().sum())
```

## Dropping the nulls and resetting index.

```python
# dropping the nulls in the dataset.
books.dropna(inplace = True)
# resetting the index.
books.reset_index(inplace = True)
# removing the newly made index column.
books.drop('index', axis = 1, inplace = True)
books.tail()
```

## Checking the information of the columns to determine which are necessary for the analysis.

```python
books.info()
```

## Dropping the unnecessary columns.

```python
books.drop(['book_id', 'goodreads_book_id', 'best_book_id', 'work_id', 'isbn', 'isbn13', 'authors', 'original_publication_year', 'title', 'language_code', 'image_url', 'small_image_url'], axis = 1, inplace = True)
books.head()
```

# "Focus your analysis on the Harry Potter book series."

## Filtering the harry potter books.

```python
# making a condition for a string containing the words "harry potter"
condition = books['original_title'].str.contains('Harry Potter')
# making a new dataframe that satisfies the condition.
harry_potter_books = books[condition]
# resetting index.
harry_potter_books.reset_index(inplace = True)
# dropping new index column.
harry_potter_books = harry_potter_books.drop('index', axis = 1)
harry_potter_books.head(10)
```

# "Find the most selling books within the Harry Potter series."

## calculating the expected number of sales as the total number of ratings given to the work.

```python
harry_potter_books['sales'] = harry_potter_books['work_ratings_count'] + harry_potter_books['work_text_reviews_count']
```

## Sorting the books by the number of sales descendingly to show the most selling books in order.

```python
harry_potter_books.sort_values(by = 'sales', ascending = False, inplace = True)
harry_potter_books.head(10)
```

## Visualizing the sales of every harry potter book.

```python
# importing necessary module.
import matplotlib.pyplot as plt
# changing plot figure.
plt.figure(figsize = (10, 10))
# making a horizontal bar graph between the book titles and their sales.
plt.barh(y = 'original_title', data = harry_potter_books, width = 'sales')
# setting a y-label.
plt.ylabel('Original Title', fontsize = 12)
# setting a x-label.
plt.xlabel('Sales', fontsize=12)
# setting a title to the plot.
plt.title('Harry Potter Sales', fontsize = 14)
# showing the plot.
plt.show()
```

# "Calculate the average rating of the Harry Potter books."

## Calculating the average rating by taking the count as weights.

```python
average_rating = (harry_potter_books.average_rating * harry_potter_books.work_ratings_count).sum() / harry_potter_books.work_ratings_count.sum()
print("Average rating of all harry potter books is:", average_rating)
```

## Visualizing the ratings of every harry potter book and their average.

```python
# changing plot figure.
plt.figure(figsize = (10, 10))
# making a horizontal bar graph between the book titles and their sales.
plt.barh(y = 'original_title', width = 'average_rating', data = harry_potter_books)
# drawing a line for the average rating mark.
plt.axvline(x = average_rating, color = 'r')
# setting a y-label.
plt.ylabel('Original Title', fontsize = 12)
# setting a x-label.
plt.xlabel('Average Rating', fontsize = 12)
# setting a title to the plot.
plt.title('Harry Potter books\' average rating', fontsize = 14)
# limiting x to be between 3.5 and 5.
plt.xlim((3.5, 5))
# tightening the plot.
plt.tight_layout()
# showing the plot.
plt.show()
```
